```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 設定 data_path
path = '/Users/liaoshousan/Desktop/3rd-ML100Days/Dataset'
f_app = os.path.join(path, 'application_train.csv')
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
import statistics
app_train['AMT_INCOME_TOTAL'].mean() , app_train['AMT_INCOME_TOTAL'].std()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(app_train['AMT_INCOME_TOTAL'],bins=50, range=(20000,1000000),rwidth=0.5)
plt.xlabel('AMT_INCOME_TOTAL',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.show()
```
